```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train_aWnotuB.csv")
df.head()
```

```python
from datetime import datetime as dt
dateobjs = [dt.strptime(dates,"%Y-%m-%d %H:%M:%S") for dates in df.DateTime.values]
```

```python
df["Year"] = [dates.year for dates in dateobjs]
df["Month"] = [dates.month for dates in dateobjs]
df["Day"] = [dates.day for dates in dateobjs]
df["DOW"] = [dates.weekday() for dates in dateobjs]
df["Hour"] = [dates.hour for dates in dateobjs]
df["timestamp"] = [dates.timestamp() for dates in dateobjs]
```

```python
df.head()
```

```python
df.to_csv("train_set_updates.csv",index=False)
```
