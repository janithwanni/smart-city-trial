```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train_aWnotuB.csv")
df.head()
```

```python
from datetime import datetime as dt
dateobjs = [dt.strptime(dates,"%Y-%m-%d %H:%M:%S") for dates in df.DateTime.values]
```

```python
df["Year"] = [dates.year for dates in dateobjs]
df["Month"] = [dates.month for dates in dateobjs]
df["Day"] = [dates.day for dates in dateobjs]
df["DOW"] = [dates.weekday() for dates in dateobjs]
df["Hour"] = [dates.hour for dates in dateobjs]
df["timestamp"] = [dates.timestamp() for dates in dateobjs]
```

```python
df.head()
```

```python
df.to_csv("train_set_updates.csv",index=False)
```

```python
def getShifted(data,steps):
    Y = data["Vehicles"]
    X = data.drop(["DateTime","ID","Vehicles"],axis=1)
    return_X = pd.DataFrame()
    return_Y = pd.DataFrame()
    XcolumnNames = X.columns
    for i in range(steps,0,-1):
        for col in XcolumnNames:
            return_X[col[:5]+"_t-"+str(i)] = X.shift(i+1-(steps))[col]
    return_Y["Vehicles_t"] = Y.shift(-(steps))
    return_X = return_X.dropna()
    return_Y = return_Y.dropna()
    return return_X,return_Y
    """
    data -> df of one junction data with vehicles and date time features
    X = pd.DataFrame()
    Y = pd.DataFrame()
    columnNames = data.columns
    resultCols = list()
    for i in range(steps,0,-1):
      for col in columnNames:
          X[col[0]+"_t-"+str(i)] = data.shift(i+1 - (steps))[col]
    for col in columnNames:
      Y[col[0]+"t"] = data.shift(-(steps))[col]
    X = X.dropna()
    Y = Y.dropna()
    return X,Y
    """
```

```python
df_dict = {junction: None for junction in df.Junction.unique()}
for junction in df.Junction.unique():
    df_dict[junction] = (getShifted(df.loc[df.Junction == junction,:],1)) # get shifted should return to here
```
